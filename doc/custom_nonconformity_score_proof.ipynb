The purpose of this documentation is to generalize the coverage guarantees of the [Predictive inference with the jackknife+](https://arxiv.org/pdf/1905.02928.pdf) methods (jacknife+ and jacknife mimax) to a larger set of nonconformity scores.

First, two function sets are definied with properties that allow to verify the same coverage properties. Then, the two theorems guaranteing coverage for the jackknife+ method and the jackknife minmax one are reviewed with these new sets of functions.

# Nonconformity scores and prediction function
In order to generalize the coverage guarantee to more nonconformity scores (not only the absolute values of residuals), let's first define two functions with some properties. The nonconformity score $f$ defined on $E^{2}$ ($E \subset \mathbb{R}$) and the prediction function $g$ defined on $F \times E$ ($F \subset \mathbb{R}$). Uppercase letters are used to follow the paper notation.

## Nonconformity scores
From an observation and a prediction, the function $f$ outputs the conformity score:
$$\begin{align}
f:E^{2} &\to F\\
(Y, \hat{Y}) &\mapsto f(Y, \hat{Y})
\end{align}$$

It verifies the following properties:
$$\begin{align}
& \forall Y_{l} < Y_{h}: f(Y_{l}, \hat{Y}) < f(Y_{h}, \hat{Y}) \\
& \forall \hat{Y}_{l} < \hat{Y}_{h}: f(Y, \hat{Y}_{l}) > f(Y, \hat{Y}_{h}) \\
& \forall i \neq j: R_{i,j} = f(Y_{i}, \hat{\mu}_{-(i,j)}(X_{i}))
\end{align}$$

## Prediction function
From a conformity score and a prediction, the function $g$ ouputs a perturbed prediction:
$$\begin{align}
g:F \times E &\to E\\
(R, \hat{Y}) &\mapsto g(R, \hat{Y})
\end{align}$$

It verifies the following properties:
$$\begin{align}
& \forall R_{l} < R_{h}: g(R_{l}, \hat{Y}) < g(R_{h}, \hat{Y}) \\
& \forall i \neq j: Y_{i} = g(R_{i,j}, \hat{\mu}_{-(i,j)}(X_{i}))
\end{align}$$

## Function relationship
The two functions must verify the following property:
$$\begin{align}
\forall \hat{Y}, \forall R: R = f(g(R, \hat{Y}), \hat{Y}) 
\end{align}$$

# The jackknife+ prediction interval
Let's derive the proof of the theorem to build [the jackknife+ prediction interval](https://arxiv.org/pdf/1905.02928.pdf) (Theorem 1, part 6, p. 18-21) with more general nonconformity scores.

## Construction of R
Instead of defining the R matrix with the absolute value of the residuals, let's assume that R is defined thanks to a nonconformity score $f$, with $R_{i,j}$ being the non conformity score of the point $i$ using a model fitted on the training plus test data, with points $i$ and $j$ removed:
$$\begin{align}
R_{i,j} &= f(Y_{i}, \hat{\mu}_{-(i,j)}(X_{i})) \\
R_{i}^{LOO} &= R_{i,n+1}
\end{align}$$

## Construction of A
According to the paper, comparison matrices A can be built. $A_{i,j}$ is the indicator for the event that, when excluding data points $i$ and $j$ from the regression, data point $i$ has a more extreme (may be higher or lower depending on the considered case, as described below) non conformity score than data point $j$. There are two cases:
- the symmetrical case when the conformity scores are assumed to be symmetrical around 0. In this case, the absolute values of the conformity scores are used to construct the A matrix.
- otherwise, the signed conformity scores are kept as they are.

### Symmetrical case
According to the paper, the comparison A matrix is defined as below:
$$A_{i,j} = \mathbb{1} \{ \lvert R_{i,j} \rvert > \lvert R_{j,i} \rvert \}$$

### Asymmetrical case
The two following definitions of comparison matrix A are considered: 
$$\begin{align}
A^{+}_{i,j} &= \mathbb{1} \{R_{i,j} > R_{j,i}\} \\
A^{-}_{i,j} &= \mathbb{1} \{R_{i,j} < R_{j,i}\} = J - A^{+}
\end{align}$$

## Strange points
As above, the definition of strange points depends on the case considered.

### Symmetrical case
According to the paper, there is a single set $S(A)=\{i \in \{1,...,n+1\}:A_{i.}\geq(1-\alpha)(n+1)\}$.

### Asymmetrical case
Instead of the single set $S(A)$, two sets should be considered: $A^{+}$ and $A^{-}$. Keeping the same definition of the set $S$ of the paper, we have:
- $S_{+}=S(A^{+})=\{i \in \{1,...,n+1\}:A^{+}_{i.}\geq(1-\frac{\alpha}{2})(n+1)\}$ for "data points with unusually large [positive] residuals", and
- $S_{-}=S(A^{-})=\{i \in \{1,...,n+1\}:A^{-}_{i.}\geq(1-\frac{\alpha}{2})(n+1) \iff (n-A^{+}_{i.})\geq(1-\frac{\alpha}{2})(n+1)\}$  for "data points with unusually large [negative] residuals".

## Step 1 and 2 are unchanged
Changing $R$ and $A$, does not change the results of steps 1 and 2 from the [proof of the original paper](https://arxiv.org/pdf/1905.02928.pdf) (Theorem 1, part 6, p. 18-21) considering absolute value of the residuals. From step 1, in the symetrical case, we get $|S(A)| \leq 2 \alpha (n+1)$. And in the asymmetrical case $|S_{+}| \leq \alpha (n+1)$ and $|S_{-}| \leq \alpha (n+1)$ since the new matrices have the same property as in the paper. From step 2, we have that the probability of the test point $n+1$ being strange inside $S_{+}$ is bounded by $\alpha$ and the probability of the test point $n+1$ being strange inside $S_{-}$ is also bounded by $\alpha$ for the asymmetrical case (and the probability of the test point $n+1$ being strange inside $S(A)$ is bounded by $2\alpha$ for the symmetrical case).

## Step 3

### Symmetrical case

Similarly to the theorem's proof, let's suppose that $Y_{n+1} \notin \hat{C}_{n,\alpha}^{jackknife+}(X_{n+1})$. Here $\hat{C}_{n,\alpha}^{jackknife+}(X_{n+1}) = [\hat{q}_{n,\alpha}^{-}\{g(- \lvert R_{i}^{LOO} \rvert, X_{n+1})\}, \hat{q}_{n,\alpha}^{+}\{g( \lvert R_{i}^{LOO} \rvert, X_{n+1})\}]$. We remind that $\hat{q}_{n,\alpha}^{+}\{v_i\}$ = the $\lceil (1 âˆ’ \alpha)(n + 1) \rceil$ -th smallest value of $v_1, ..., v_n$ and $\hat{q}_{n,\alpha}^{-}\{v_i\}$ = the $\lfloor \alpha(n + 1) \rfloor$ -th smallest value of $v_1, ..., v_n$. This means either:
$$ Y_{n+1} > \hat{q}_{n,\alpha}^{+}\{ g( \lvert R_{i}^{LOO} \rvert , \hat{\mu}_{-i}(X_{n+1})) \}$$
which implies that $Y_{n+1} > g( \lvert R_{i}^{LOO} \rvert , \hat{\mu}_{-i}(X_{n+1}))$ for at least $(1-\alpha)(n+1)$ many indices $i \in \{1,...,n\}$ since $g$ is an increasing function regarding the first variable,.
$$\begin{align}
(1-\alpha)(n+1) & \leq \sum_{i=1}^{n} \mathbb{1}\{ Y_{n+1} > g( \lvert R_{i}^{LOO} \rvert , \hat{\mu}_{-i}(X_{n+1})) \}\\
& = \sum_{i=1}^{n} \mathbb{1}\{ f( Y_{n+1} , \hat{\mu}_{-i}(X_{n+1}) ) > f( g( \lvert R_{i}^{LOO} \rvert, \hat{\mu}_{-i}(X_{n+1})), \hat{\mu}_{-i}(X_{n+1})) \} \\
& = \sum_{i=1}^{n} \mathbb{1}\{ \lvert f( g(R_{n+1,i}, \hat{\mu}_{-i}(X_{n+1})) , \hat{\mu}_{-i}(X_{n+1}) ) \rvert > \lvert R_{i}^{LOO} \rvert \} \\
& = \sum_{i=1}^{n} \mathbb{1}\{ R_{n+1,j} > \lvert R_{i}^{LOO} \rvert \} \\
& \leq \sum_{i=1}^{n} \mathbb{1}\{ \lvert R_{n+1,i} \rvert > \lvert R_{i}^{LOO} \rvert \}  = \sum_{i=1}^{n} A_{n+1,i}
\end{align}$$

Or otherwise:
$$ Y_{n+1} < \hat{q}_{n,\alpha}^{-}\{ g( - \lvert R_{i}^{LOO} \rvert , \hat{\mu}_{-i}(X_{n+1})) \}$$
which implies that $Y_{n+1} < g( - \lvert R_{i}^{LOO} \rvert , \hat{\mu}_{-i}(X_{n+1}))$ for at least $(1-\alpha)(n+1)$ many indices $i \in \{1,...,n\}$.
$$\begin{align}
(1-\alpha)(n+1) & \leq \sum_{i=1}^{n} \mathbb{1}\{ Y_{n+1} < g( - \lvert R_{i}^{LOO} \rvert , \hat{\mu}_{-i}(X_{n+1})) \}\\
& = \sum_{i=1}^{n} \mathbb{1}\{ f( Y_{n+1} , \hat{\mu}_{-i}(X_{n+1}) ) < f( g( - \lvert R_{i}^{LOO} \rvert, \hat{\mu}_{-i}(X_{n+1})), \hat{\mu}_{-i}(X_{n+1})) \} \\
& = \sum_{i=1}^{n} \mathbb{1}\{ f( g(R_{n+1,i}, \hat{\mu}_{-j}(X_{n+1})) , \hat{\mu}_{-i}(X_{n+1}) ) < - \lvert R_{i}^{LOO} \rvert \} \\
& = \sum_{i=1}^{n} \mathbb{1}\{ R_{n+1,i} < - \lvert R_{i}^{LOO} \rvert \} \\
& \leq \sum_{i=1}^{n} \mathbb{1}\{ \lvert R_{n+1,i} \rvert > \lvert R_{i}^{LOO} \rvert \}  = \sum_{i=1}^{n} A_{n+1,j}
\end{align}$$

Therefore, in either case, $n+1 \in S(A)$, that is, point $n+1$ is a strange point with a high absolute conformity score.

Thus:
$$\mathbb{P}\{ Y_{n+1} \notin \hat{C}_{n,\alpha}^{jackknife+}(X_{n+1}) \} \leq 2\alpha $$
and:
$$\mathbb{P}\{ Y_{n+1} \in \hat{C}_{n,\alpha}^{jackknife+}(X_{n+1}) \} \geq 1-2\alpha $$

### Asymmetrical case

Similarly to the theorem's proof, let's suppose that $Y_{n+1} \notin \hat{C}_{n,\alpha}^{jackknife+}(X_{n+1})$.  Here $\hat{C}_{n,\alpha}^{jackknife+}(X_{n+1}) = [\hat{q}_{n,\frac{\alpha}{2}}^{-}\{g(R_{i}^{LOO}, X_{n+1})\}, \hat{q}_{n,\frac{\alpha}{2}}^{+}\{g(R_{i}^{LOO}, X_{n+1})\}]$This means either:
$$Y_{n+1}>\hat{q}_{n,\frac{\alpha}{2}}^{+}\{g(R_{i}^{LOO}, \hat{\mu}_{-i}(X_{n+1}))\}$$
which implies that $Y_{n+1} > g(R_{i}^{LOO}, \hat{\mu}_{-i}(X_{n+1}))$ for at least $(1-\frac{\alpha}{2})(n+1)$ many indices $j \in \{1,...,n\}$ since $g$ is an increasing function regarding the first variable, or otherwise:
$$Y_{n+1}<\hat{q}_{n,\frac{\alpha}{2}}^{-}\{g(R_{i}^{LOO}, \hat{\mu}_{-i}(X_{n+1}))\}$$
which implies that $Y_{n+1} < g(R_{i}^{LOO}, \hat{\mu}_{-i}(X_{n+1}))$ for at least $(1-\frac{\alpha}{2})(n+1)$ many indices $j \in \{1,...,n\}$.

Here $R_{j}^{LOO}$ can be positive or negative so it is added in both cases. Besides, $\frac{\alpha}{2}$ is used instead of $\alpha$ since to get the $(1-\alpha)$ confidence interval, we look for an upper bound on the confidence interval so that $\frac{\alpha}{2}$ of the residuals are above and a lower bound on the confidence interval so that $\frac{\alpha}{2}$ of the residuals are below.

Let's consider the first case $ Y_{n+1} > \hat{q}_{n,\frac{\alpha}{2}}^{+}\{ g(R_{j}^{LOO}, \hat{\mu}_{-j}(X_{n+1})) \} $. Then, thanks to the properties of $f$ and $g$:
$$\begin{align}
(1-\frac{\alpha}{2})(n+1) & \leq \sum_{j=1}^{n} \mathbb{1}\{Y_{n+1} > g(R_{j}^{LOO}, \hat{\mu}_{-j}(X_{n+1})) \} \\
&= \sum_{j=1}^{n} \mathbb{1}\{ f(Y_{n+1}, \hat{\mu}_{-j}(X_{n+1})) > f(g(R_{j}^{LOO}, \hat{\mu}_{-j}(X_{n+1})), \hat{\mu}_{-i}(X_{n+1})) \} \\
&= \sum_{j=1}^{n} \mathbb{1}\{ f(g(R_{n+1,j}, \hat{\mu}_{-j}(X_{n+1})), \hat{\mu}_{-j}(X_{n+1})) > R_{j}^{LOO} \} \\
&= \sum_{j=1}^{n} \mathbb{1}\{ R_{n+1,j} > R_{j,n+1}\} = \sum_{j=1}^{n} A^{+}_{n+1,j}
\end{align}$$
and therefore $n+1 \in S_{+}$, that is, point $n+1$ is a strange point with a high positive residual.

Similarly, we can conclude the same for strange point with a high negative residual.
$$\begin{align}
(1-\frac{\alpha}{2})(n+1) & \leq \sum_{j=1}^{n} \mathbb{1}\{Y_{n+1} < g(R_{j}^{LOO}, \hat{\mu}_{-j}(X_{n+1})) \} \\
&= \sum_{j=1}^{n} \mathbb{1}\{ f(Y_{n+1}, \hat{\mu}_{-j}(X_{n+1})) < f(g(R_{j}^{LOO}, \hat{\mu}_{-j}(X_{n+1})), \hat{\mu}_{-j}(X_{n+1})) \} \\
&= \sum_{j=1}^{n} \mathbb{1}\{ f(g(R_{n+1,j}, \hat{\mu}_{-j}(X_{n+1})), \hat{\mu}_{-j}(X_{n+1})) < R_{j}^{LOO}\} \\
&= \sum_{j=1}^{n} \mathbb{1}\{R_{n+1,j} < R_{j,n+1}\} = \sum_{j=1}^{n} A^{-}_{n+1,j} = n - \sum_{j=1}^{n} A^{+}_{n+1,j}
\end{align}$$
and therefore $n+1 \in S_{-}$, that is, point $n+1$ is a strange point with a high negative residual.

Combining with step 2, we have:
$$\begin{align}
\mathbb{P}\{Y_{n+1}>\hat{q}_{n,\frac{\alpha}{2}}^{+}\{ g(R_{j}^{LOO}, \hat{\mu}_{-j}(X_{n+1})) \} &\leq \alpha \\
\mathbb{P}\{Y_{n+1}<\hat{q}_{n,\frac{\alpha}{2}}^{-}\{ g(R_{j}^{LOO}, \hat{\mu}_{-j}(X_{n+1})) \} &\leq \alpha
\end{align}$$

Since any point cannot be simultaneously in $S_{+}$ and $S_{-}$ otherwise $\alpha \geq 1$ (because $A^{+}_{i.}\geq(1-\frac{\alpha}{2})(n+1)$ and $(n-A^{+}_{i.})\geq(1-\frac{\alpha}{2})(n+1)$ => $n\geq2(1-\frac{\alpha}{2})(n+1) \iff \alpha \geq \frac{n+2}{n+1}$), a point is at most in one of the two sets.

$$\begin{align}
\mathbb{P}\{ Y_{n+1} \in \hat{C}_{n,\alpha}^{jackknife+}(X_{n+1}) \}
&=  1 - \mathbb{P}\{ Y_{n+1} > \hat{q}_{n,\frac{\alpha}{2}}^{+}\{ g(R_{j}^{LOO}, \hat{\mu}_{-j}(X_{n+1})) \} - \mathbb{P}\{ Y_{n+1} < \hat{q}_{n,\frac{\alpha}{2}}^{-}\{ g(R_{j}^{LOO}, \hat{\mu}_{-j}(X_{n+1})) \}\\
&\geq 1-2\alpha
\end{align}$$

# The jackknife-minmax prediction interval
let's derive the proof of the theorem to build [the jackknife-minmax prediction interval](https://arxiv.org/pdf/1905.02928.pdf) (Theorem 3, part B.1, p. 31-32) with more general nonconformity scores.

## Construction of R
As above, instead of defining the R matrix with the absolute residual, let assume that R is defined thanks to a nonconformity score:
$$\begin{align}
R_{i,j} &= f(Y_{i}, \hat{\mu}_{-(i,j)}(X_{i})) \\
R_{i}^{LOO} &= R_{i,n+1}
\end{align}$$

## Construction of A
As above, there are two cases:
- the symmetrical case when the conformity scores are assumed to be symmetrical around 0. In this case, the absolute values of the conformity scores are used to construct the A matrix.
- otherwise, the signed conformity scores are kept as they are.

### Symmetrical case
According to the paper, the A matrix is defined as below:
$$A_{i,j} = \mathbb{1} \{ \underset{j'}{\min} \lvert R_{i,j'} \rvert > \lvert R_{j,i} \rvert \}$$

### Asymmetrical case
The two following definitions of matrix A are considered: 
$$\begin{align}
A^{+}_{i,j} &= \mathbb{1} \{\underset{j'}{\min} R_{i,j'} > R_{j,i}\} \\
A^{-}_{i,j} &= \mathbb{1} \{\underset{j'}{\max} R_{i,j'} < R_{j,i}\} = \mathbb{1} \{\underset{j'}{\min} -R_{i,j'} > -R_{j,i}\}
\end{align}$$

## Strange points
As above, the definition of strange points depends on the case considered.

### Symmetrical case
According to the paper, there is a single $S(A)$ set $S(A)=\{i \in \{1,...,n+1\}:A_{i.}\geq(1-\alpha)(n+1)\}$.

### Asymmetrical case
Instead of the single set $S(A)$, two sets should be considered: $A^{+}$ and $A^{-}$. Keeping the same definition of the set $S$ of the paper, we have:
- $S_{+}=S(A^{+})=\{i \in \{1,...,n+1\}:A^{+}_{i.}\geq(1-\frac{\alpha}{2})(n+1)\}$ for "data points with unusually large [positive] residuals", and
- $S_{-}=S(A^{-})=\{i \in \{1,...,n+1\}:A^{-}_{i.}\geq(1-\frac{\alpha}{2})(n+1) \iff (n-A^{+}_{i.})\geq(1-\frac{\alpha}{2})(n+1)\}$  for "data points with unusually large [negative] residuals".

## Step 1 and 2 are unchanged
Changing $R$ and $A$, does not change the results of steps 1 and 2  from the [proof of the original paper](https://arxiv.org/pdf/1905.02928.pdf) (Theorem 1, part 6, p. 18-21) considering absolute value of the residuals. From step 1, in the symetrical case, we get $|S(A)| \leq \alpha (n+1)$. And in the asymmetrical case $|S_{+}| \leq \frac{\alpha}{2} (n+1)$ and $|S_{-}| \leq \frac{\alpha}{2} (n+1)$ since the new matrices have the same property as in the paper. From step 2, we have that the probability of the test point $n+1$ being strange inside $S_{+}$ is bounded by $\frac{\alpha}{2}$ and the probability of the test point $n+1$ being strange inside $S_{-}$ is also bounded by $\frac{\alpha}{2}$ for the asymmetrical case (and the probability of the test point $n+1$ being strange inside $S(A)$ is bounded by $\alpha$ for the symmetrical case).

## Step 3

### Symmetrical case
Similarly to the theorem's proof, let's suppose that $Y_{n+1} \notin \hat{C}_{n,\alpha}^{jack-mm}(X_{n+1})$. Here $\hat{C}_{n,\alpha}^{jack-mm}(X_{n+1}) = [g( \hat{q}_{n,\alpha}^{-}\{ - \lvert R_{j}^{LOO} \rvert \} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})) , g( \hat{q}_{n,\alpha^{+}}\{ \lvert R_{j}^{LOO} \rvert \} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1}))]$. This means either:
$$ Y_{n+1} > g( \hat{q}_{n,\alpha^{+}}\{ \lvert R_{j}^{LOO} \rvert \} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) $$
which implies that $ Y_{n+1} > g( \lvert R_{j}^{LOO} \rvert , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) $ for at least $(1-\alpha)(n+1)$ many indices $j \in \{1,...,n\}$ since $g$ is an increasing function regarding the first variable,.
$$\begin{align}
(1-\alpha)(n+1) & \leq \sum_{j=1}^{n} \mathbb{1}\{ Y_{n+1} > g( \lvert R_{j}^{LOO} \rvert , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) \}\\
& = \sum_{j=1}^{n} \mathbb{1}\{ f( Y_{n+1} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1}) ) > f( g( \lvert R_{j}^{LOO} \rvert, \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1}) ), \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1}) ) \} \\
& = \sum_{j=1}^{n} \mathbb{1}\{ \underset{i}{\min} f( Y_{n+1} , \hat{\mu}_{-i}(X_{n+1}) ) = f( Y_{n+1} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1}) ) > \lvert R_{j}^{LOO} \rvert \} \\
& = \sum_{j=1}^{n} \mathbb{1}\{ \underset{i}{\min}R_{n+1,i} > \lvert R_{j}^{LOO} \rvert \} \\
& \leq \sum_{j=1}^{n} \mathbb{1}\{ \underset{i}{\min} \lvert R_{n+1,i} \rvert > \lvert R_{j}^{LOO} \rvert \}  = \sum_{j=1}^{n} A_{n+1,j}
\end{align}$$

Or otherwise:
$$ Y_{n+1} < g( \hat{q}_{n,\alpha}^{-}\{ - \lvert R_{j}^{LOO} \rvert \} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})) $$
which implies that $ Y_{n+1} < g( - \lvert R_{j}^{LOO} \rvert , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})) $ for at least $(1-\alpha)(n+1)$ many indices $j \in \{1,...,n\}$.
$$\begin{align}
(1-\alpha)(n+1) & \leq \sum_{j=1}^{n} \mathbb{1}\{ Y_{n+1} < g( - \lvert R_{j}^{LOO} \rvert , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})) \}\\
& = \sum_{j=1}^{n} \mathbb{1}\{ f( Y_{n+1} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1}) ) < f( g( - \lvert R_{j}^{LOO} \rvert, \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1}) ), \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1}) ) \} \\
& = \sum_{j=1}^{n} \mathbb{1}\{ \underset{i}{\max} f( Y_{n+1} , \hat{\mu}_{-i}(X_{n+1}) ) = f( Y_{n+1} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1}) ) < - \lvert R_{j}^{LOO} \rvert \} \\
& = \sum_{j=1}^{n} \mathbb{1}\{ \underset{i}{\max}R_{n+1,i} < - \lvert R_{j}^{LOO} \rvert \} \\
& \leq \sum_{j=1}^{n} \mathbb{1}\{ - \underset{i}{\min} \lvert R_{n+1,i} \rvert < -\lvert R_{j}^{LOO} \rvert \}  = \sum_{j=1}^{n} A_{n+1,j}
\end{align}$$

Therefore, in either case, $n+1 \in S(A)$, that is, point $n+1$ is a strange point with a high absolute conformity score.

Thus:
$$\mathbb{P}\{ Y_{n+1} \notin \hat{C}_{n,\alpha}^{jackknife+}(X_{n+1}) \} \leq \alpha $$
and:
$$\mathbb{P}\{ Y_{n+1} \in \hat{C}_{n,\alpha}^{jackknife+}(X_{n+1}) \} \geq 1-\alpha $$

### Asymmetrical case

Similarly to the theorem's proof, let's suppose that $Y_{n+1} \notin \hat{C}_{n,\alpha}^{jack-mm}(X_{n+1})$. Here $\hat{C}_{n,\alpha}^{jack-mm}(X_{n+1}) = [g( \hat{q}_{n,\frac{\alpha}{2}}^{-}\{ R_{j}^{LOO} \} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})) , g( \hat{q}_{n,\frac{\alpha}{2}}^{+}\{ R_{j}^{LOO} \} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1}))]$. This means either:
$$ Y_{n+1} > g( \hat{q}_{n,\frac{\alpha}{2}}^{+}\{ R_{j}^{LOO} \} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) $$
which implies that $Y_{n+1} > g( R_{j}^{LOO} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) $ for at least $(1-\frac{\alpha}{2})(n+1)$ many indices $j \in \{1,...,n\}$ since $g$ is an increasing function regarding the first variable, or otherwise:
$$Y_{n+1} < g( \hat{q}_{n,\frac{\alpha}{2}}^{-}\{ R_{j}^{LOO} \} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1}))$$
which implies that $Y_{n+1} < g( R_{j}^{LOO} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1}))$ for at least $(1-\frac{\alpha}{2})(n+1)$ many indices $j \in \{1,...,n\}$.

Here $R_{j}^{LOO}$ can be positive or negative so it is added in both cases. Besides, $\frac{\alpha}{2}$ is used instead of $\alpha$ since to get the $(1-\alpha)$ confidence interval, we look for an upper bound on the confidence interval so that $\frac{\alpha}{2}$ of the residuals are above and a lower bound on the confidence interval so that $\frac{\alpha}{2}$ of the residuals are below.

Let's consider the first case $ Y_{n+1} > g( R_{j}^{LOO} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) $. We have:
$$\begin{align}
(1-\frac{\alpha}{2})(n+1) & \leq \sum_{j=1}^{n} \mathbb{1}\{
Y_{n+1} > g( R_{j}^{LOO} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1}))
\} \\
&= \sum_{j=1}^{n} \mathbb{1}\{
f(Y_{n+1}, \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) > f( g( R_{j}^{LOO} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})), \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1}))
\} \\
&= \sum_{j=1}^{n} \mathbb{1}\{
\underset{i}{\min}f(Y_{n+1}, \hat{\mu}_{-i}(X_{n+1})) = f(Y_{n+1}, \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) > R_{j}^{LOO}
\} \\
&= \sum_{j=1}^{n} \mathbb{1}\{
\underset{i}{\min} R_{n+1,i} > R_{j,n+1}
\} = \sum_{j=1}^{n} A^{+}_{n+1,j}
\end{align}$$
and therefore $n+1 \in S_{+}$, that is, point $n+1$ is a strange point with a high positive residual.

Similarly, we can conclude the same for strange point with a high negative residual.
$$\begin{align}
(1-\frac{\alpha}{2})(n+1) & \leq \sum_{j=1}^{n} \mathbb{1}\{
Y_{n+1} < g( R_{j}^{LOO} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1}))
\} \\
&= \sum_{j=1}^{n} \mathbb{1}\{
f(Y_{n+1}, \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})) < f( g( R_{j}^{LOO} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})), \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1}))
\} \\
&= \sum_{j=1}^{n} \mathbb{1}\{
\underset{i}{\max}f(Y_{n+1}, \hat{\mu}_{-i}(X_{n+1})) = f(Y_{n+1}, \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})) < R_{j}^{LOO}
\} \\
&= \sum_{j=1}^{n} \mathbb{1}\{
\underset{i}{\max} R_{n+1,i} < R_{j,n+1}
\} = \sum_{j=1}^{n} A^{-}_{n+1,j}
\end{align}$$
and therefore $n+1 \in S_{-}$, that is, point $n+1$ is a strange point with a high negative residual.

Combining with step 2, we have:
$$\begin{align}
\mathbb{P}\{
Y_{n+1} > g( \hat{q}_{n,\frac{\alpha}{2}}^{+}\{ R_{j}^{LOO} \} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) \} &\leq \frac{\alpha}{2} \\
\mathbb{P}\{
Y_{n+1} < g( \hat{q}_{n,\frac{\alpha}{2}}^{-}\{ R_{j}^{LOO} \} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})) \} &\leq \frac{\alpha}{2}
\end{align}$$

Since any point cannot be simultaneously in $S_{+}$ and $S_{-}$ because the boundaries are even larger than in the jackknife+ case, we finally have:

$$\begin{align}
\mathbb{P}\{Y_{n+1} \in \hat{C}_{n,\alpha}^{jack-mm}(X_{n+1})\}
&=  1 - \mathbb{P}\{
Y_{n+1} > g( \hat{q}_{n,\frac{\alpha}{2}}^{+}\{ R_{j}^{LOO} \} , \underset{i}{\max}\hat{\mu}_{-i}(X_{n+1})) \}  - \mathbb{P}\{
Y_{n+1} < g( \hat{q}_{n,\frac{\alpha}{2}}^{-}\{ R_{j}^{LOO} \} , \underset{i}{\min}\hat{\mu}_{-i}(X_{n+1})) \}\\
&\geq 1-\alpha
\end{align}$$

# Examples

## Absolute residual conformity score
The absolute residual conformity score, used in the paper, is among the set of nonconformity scores thanks to the following functons and by considering the symmetrical case.
$$\begin{align}
E &= \mathbb{R} \\
F &= \mathbb{R} \\
f(Y, \hat{Y}) &= Y - \hat{Y} \\
g(R, \hat{Y}) &= \hat{Y} + R
\end{align}$$

## Gamma conformity score
The Gamma conformity score is among the set of nonconformity scores thanks to the following functons and by considering the asymmetrical case.
$$\begin{align}
E &= \mathbb{R}^{+*} \\
F &= (-1;+\infty) \\
f(Y, \hat{Y}) &= \frac{Y}{\hat{Y}} - 1 \\
g(R, \hat{Y}) &= \hat{Y}(1+R)
\end{align}$$

## Tweedie conformity score
The Tweedie conformity score is among the set of nonconformity scores thanks to the following functons and by considering the asymmetrical case.
$$\begin{align}
E &= \mathbb{R}^{+*} \\
F &= \mathbb{R} \\
p &\in (0;2) \\
f(Y, \hat{Y}) &= \frac{Y-\hat{Y}}{\hat{Y}^{\frac{p}{2}}} \\
g(R, \hat{Y}) &= \hat{Y}(1+R \hat{Y}^{\frac{p}{2}-1})
\end{align}$$
